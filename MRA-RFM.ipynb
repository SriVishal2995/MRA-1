```python
df_year = data.groupby(by=['QUANTITYORDERED', 'ORDERLINENUMBER']).agg({'SALES': lambda x:x.sum() })
```

```python
df_year.head()
```

```python
df_recency = data.groupby(by='ORDERLINENUMBER',
                        as_index=False)['ORDERDATE'].max()
df_recency.columns = ['ORDERLINENUMBER', 'DAYS_SINCE_LASTORDER']
recent_date = df_recency['DAYS_SINCE_LASTORDER'].max()
df_recency['Recency'] = df_recency['DAYS_SINCE_LASTORDER'].apply(
    lambda x: (recent_date - x).days)
df_recency.head()
```

```python
frequency_df = data.drop_duplicates().groupby(
    by=['ORDERLINENUMBER'], as_index=False)['ORDERDATE'].count()
frequency_df.columns = ['ORDERLINENUMBER', 'Frequency']
frequency_df.head()
```

```python
data['Total'] = data['SALES']*data['QUANTITYORDERED']
monetary_df = data.groupby(by='ORDERLINENUMBER', as_index=False)['Total'].sum()
monetary_df.columns = ['ORDERLINENUMBER', 'Monetary']
monetary_df.head()
```
